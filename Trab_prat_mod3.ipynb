# 1. Importar Bibliotecas

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from scipy.stats import pearsonr
from scipy import stats
```

# 2. Ler os dados

```python
df = pd.read_csv('ia_dataset_igti.xls - breast-cancer.csv', sep=',', encoding='UTF-8')
df.head()
```

# 3. Entender os dados

```python
df.info()
```

# 4. Tratamento dos Dados

## 4.1. Substituir a vírgula por ponto

```python
df['radius'] = df['radius'].str.replace(',','.')
df['texture'] = df['texture'].str.replace(',','.')
df['perimeter'] = df['perimeter'].str.replace(',','.')
df['area'] = df['area'].str.replace(',','.')
df['smoothness'] = df['smoothness'].str.replace(',','.')
df['compactness'] = df['compactness'].str.replace(',','.')
df['concavity'] = df['concavity'].str.replace(',','.')
df['concave'] = df['concave'].str.replace(',','.')
df['symmetry'] = df['symmetry'].str.replace(',','.')
df['fractal_dimension'] = df['fractal_dimension'].str.replace(',','.')
```

## 4.2. Converter os dados para float

```python
df = df.astype('float')
```

```python
df.info()
```

```python
# Moda
df.mode()
```

```python
# Desvio-Padrão
df.std()
```
